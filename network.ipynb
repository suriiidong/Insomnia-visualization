```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import plotly.express as px
import plotly 
import warnings
warnings.filterwarnings("ignore")
import re
```

```python
tns=pd.read_csv('data/TeamNoSleep.csv')
tns = tns[tns['language']=='en']
ambien=pd.read_csv('data/ambien.csv')
ambien = ambien[ambien['language']=='en']
cant_sleep=pd.read_csv('data/cant_sleep.csv')
cant_sleep = cant_sleep[cant_sleep['language']=='en']
Insomnia=pd.read_csv('data/Insomnia.csv')
Insomnia = Insomnia[Insomnia['language']=='en']
melatonin=pd.read_csv('data/melatonin.csv')
melatonin = melatonin[melatonin['language']=='en']
s_a=pd.read_csv('data/sleep_aid.csv')
s_a = s_a[s_a['language']=='en']
```

```python
df_all = pd.concat([tns, ambien, cant_sleep, Insomnia, melatonin, s_a])
df_all = df_all[['Content', 'userid']]
df_all
```

```python
df_all['uid']=0
df_all
```

```python
dict_uid={}
label=0
for i in range (len(df_all)):
    uid=df_all.iloc[i,1]
    if uid in dict_uid.keys():
        df_all.iloc[i,2]=dict_uid[uid]
    else:
        dict_uid[uid]=label
        df_all.iloc[i,2]=label
        label+=1
df_all
```

```python
df_all=df_all[['Content','uid']]
df=pd.DataFrame(df_all['uid'].value_counts(ascending=False))
df=df.reset_index()
df=df[df['uid']>=5]
df=df.rename(columns={'index':'uid','uid':'count'})
df
```

```python
uid_m5_l=list(df['uid'])
def find_active(x):
    if x in uid_m5_l:
        return "active"
    else:
        return "not"
```

```python
df_all['activeornot']=df_all['uid'].apply(lambda x: find_active(x))
df_all
```

```python
df_active=df_all[df_all['activeornot']=='active']
df_active=df_active[['Content','uid']]
df_active
```

```python
# 'love','relax','stress','suffer','damn'
def find_love(x):
    if 'love' in x:
        j=1
    else:
        j=0
    return j

def find_relax(x):
    if 'relax' in x:
        j=1
    else:
        j=0
    return j

def find_stress(x):
    if 'stress' in x:
        j=1
    else:
        j=0
    return j

def find_suffer(x):
    if 'suffer' in x:
        j=1
    else:
        j=0
    return j

def find_damn(x):
    if 'damn' in x:
        j=1
    else:
        j=0
    return j
```

```python
df_active['love']=df_active['Content'].apply(lambda x: find_love(x))
df_active['relax']=df_active['Content'].apply(lambda x: find_relax(x))
df_active['stress']=df_active['Content'].apply(lambda x: find_stress(x))
df_active['suffer']=df_active['Content'].apply(lambda x: find_suffer(x))
df_active['damn']=df_active['Content'].apply(lambda x: find_damn(x))
df_active['check']=df_active['love']+df_active['relax']+df_active['stress']+df_active['suffer']+df_active['damn']
df_active
```

```python
df_active=df_active[df_active['check']>0]
df_active
```

```python
df_tem=df_active[['uid','love','relax','stress','suffer','damn']]
df_tem=df_tem.groupby(['uid']).sum()
df_tem=df_tem.reset_index()
df_tem['check']=df_tem['love']+df_tem['relax']+df_tem['stress']+df_tem['suffer']+df_tem['damn']
df_tem=df_tem.sort_values(by='check', ascending=False)
df_tem=df_tem[df_tem['check']>20]
df_tem
```

```python
df_tem.to_csv('network.csv')
```

```python
def convert(x):
    if x=='love':
        a='G1'
    elif x=='stress':
        a='G2'
    elif x=='damn':
        a='G3'
    elif x=='relax':
        a='G4'
    else:
        a='G5'
    return a
```

```python
#after processing by excel
got_data = pd.read_csv('data/network_2.csv')
got_data['userid']=got_data['source'].apply(lambda x: str(x))
got_data['group']=got_data['target'].apply(lambda x: convert(x))
got_data
```

```python
got_data=got_data[['target','value','userid','group']]
tem=got_data.groupby(['userid']).sum(['value'])
tem=tem.reset_index()
tem
```

```python
from pyvis.network import Network
import pandas as pd

got_net = Network(height="750px", width="80%", neighborhood_highlight=True, bgcolor="#213150", font_color="white")

# set the physics layout of the network
got_net.barnes_hut(gravity=-100000, central_gravity=7, spring_length=5,spring_strength=0.01)

sources = got_data['userid']
targets = got_data['target']
weights = got_data['value']
group = got_data['group']
user_sum = got_data['user_t_s']


edge_data = zip(sources, targets, weights, group, user_sum)
for e in edge_data:
    src = e[0]    
    dst = e[1]
    w = e[2]
    g = e[3]
    u = e[4]
    got_net.add_node(src, src, size=50, title=str(u))
    got_net.add_node(dst, dst, shape='dot', group=g, size=100, title=dst)
    got_net.add_edge(src, dst, value=w, title=w)
    
neighbor_map = got_net.get_adj_list()

got_net.show_buttons()
got_net.show("network_try.html")
```
